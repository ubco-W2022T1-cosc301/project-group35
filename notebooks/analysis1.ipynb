# Are Planets in multiple-star systems habitable?

A planet's habitability can be determined through it's characteristics, enviromental stability, atmosphere, gravitation interactions within it's orbit(s), ability to deflect or be durible towards minor bodies, and have liquid water. Often, due to the complications of gravitational interactions with planets and multiple-star systems it is less likely for the possibility of habitable planets to occur / exist. 

Other factors being - Mass, orbit properties, Temperature

This can be determined by testing the currently known theoretical requirements for life to exist on planets in multiple-star systems.
