## Exoplanets Discovered by year

This is the count of exoplanets discovered per year and the exponential increase can be seen clearly here that demonstrates the technological advances we have produced.

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../data/raw/DataOriginal.csv")
ax = sns.countplot(data = df, x = 'Discovery Year', palette = 'rocket')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right", fontsize=6)
ax.set_xlabel("Discovery Year", fontsize=10)
ax.set_ylabel("Count of Exoplanets discovered", fontsize=10)
ax.set_title("Exoplanets Discovered per Year", fontsize=10)
plt.savefig('Kevin1')
```

## Orbital Periods Days

This is the graph of the orbital period in days of exoplanets after the exclusion of several outliers, this shows how the trend as mass increases the orbital period tends to decrease and shows how Earth-sized planets tend to be closer to the sun or its star

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../data/raw/DataOriginal.csv")
df.dropna(subset=["Orbital Period Days"])
ax = sns.scatterplot(data=df[-(df['Orbital Period Days'] > 69000)], x = 'Mass', y = 'Orbital Period Days')
ax.set_title("Mass in relation to Orbital Period")
plt.savefig('kevin3')
```

```python
import pandas as pd
def load_and_process(url_or_path_to_csv_file):

        df1 = (
              pd.read_csv(url_or_path_to_csv_file)    
              .drop(['No.','Insolation Flux', 'Equilibrium Temperature'], axis=1)
             .rename(columns={'Planet Host': 'Host Star of Planet', 'Orbital Period Days':'Orbital Period in Days'}) 
            .dropna(axis=0)
            )
        
        return df1
```
