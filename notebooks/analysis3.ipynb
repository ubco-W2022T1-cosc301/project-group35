## How far are these exoplanets from us?

Currently, the feasibility of in-depth research and information gathering is higher the closer these exo-planets are to us. This data is measured in light years and illustrates the exo-planets within 10 light years. The exo-planets in the table, while not being the most likely to sustain life, are the closest too us therefore have the highest possiblity of scientific research and analysis.

```python
import pandas as pd 
df = pd.read_csv("../data/raw/DataOriginal.csv")


DistanceTest = ['Planet Name','Num Planets','Mass','Distance']
DF = df[DistanceTest]

Count=DF.sort_values(['Distance'], ascending = True)
Count=pd.DataFrame(Count)
Count=Count[0:11]
Count
```

## The Discovery of Exo-planets

The discovery of exo-planets increased exponentially through the years with advances in science and technology. The number of found exo-planets increased along with the distance of exo-planets. New and improved technology enabled us to see further away than ever before.

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../data/raw/DataOriginal.csv")


DistanceTest = ['Planet Name','Num Planets','Discovery Year','Distance']
DF = df[DistanceTest]

Count=DF.sort_values(['Discovery Year'], ascending = True)
Count=pd.DataFrame(Count)
Count=Count[0:1000]
Count

plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=0.5)

Count_g = sns.PairGrid(Count,
                      x_vars=["Discovery Year"],
                      y_vars=["Distance"])

Count_g.map(plt.scatter)
```
