#What planets have the highest orbital period days
To determine the habitability of a planet in any solar system, we have to first be sure of the orbital properties, mass, possibility of liquid water and so on.



```python
import pandas as pd 
df = pd.read_csv("../data/raw/DataOriginal.csv")


Orbital_days = ['Planet Name','Num Stars','Num Planets','Orbital Period Days'] 
df9 = df[Orbital_days]


df11=df9.sort_values(['Orbital Period Days'], ascending = False)
df11.dropna(subset=['Orbital Period Days'])
df16=df11.head(10)

df16.plot.bar(x="Planet Name", y="Orbital Period Days",rot=70, title="Bar Chart displaying the Orbital Period Days each planet has")





```

#What methods were used to discover the planets?
This analysis will help us discover the most popular method of discovering planets

```python
import pandas as pd
df = pd.read_csv("../data/raw/DataOriginal.csv")


discover1 = ['Discovery Method','Planet Name']
df19 = df[discover1]

df19.head()

df20 = df19.groupby('Discovery Method').count()
df21=df20.rename(columns={'Planet Name':'Num Times'})
df21.reset_index(inplace=True)
df21

df21.plot.bar(x="Discovery Method", y="Num Times",rot=70, title="Bar Chart displaying the Number of Times a method was used to discover a planet")

```

```python
def change_and_drop(url_or_path_to_csv_file):

        df5 = (
              pd.read_csv(url_or_path_to_csv_file)    
              .drop(['Insolation Flux', 'Equilibrium Temperature', 'Eccentricity','Spectral Type','Gaia Magnitude'], axis=1)  
            .dropna(axis=0)
            .sort_values(['Discovery Year'], ascending=False)
            )
        
        return df5
```
